```python
##Write a Python program to check whether all numbers in a list are positive.
x=[1,3,7,9,9,4,6]
all_positives=True
for i in x:
    if i<=0:
        all_positives=False
        break
if all_positives:
    print("yes")
else:
  print("no")
```

```python
##Write a Python program to check whether all numbers in a list are positive using all()
x=[1,3,6,8,8,8]
if all(i>0 for i in x):
    print("yes")
else:
    print("no")
```

```python
##Write a Python program to check whether a list contains any positive values using any().
x=[-1,-3,-4,-4,-4]
if any(i>0 for i in x):
    print("yes")
else:
    print("no")
```

```python
#If you pass None as the first argument, Python will automatically filter out any elements that are evaluated as False (like 0, Empty strings, None, or False itself), leaving only the "truthy" values:
data = [True, False, 1, 0, "Python", ""]
truthy_data = list(filter(None, data))

print(truthy_data)  # Output: [True, 1, 'Python']
```
